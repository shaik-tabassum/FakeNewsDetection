```python
import pandas as pd
import numpy as np
import itertools
```

```python
df=pd.read_csv("news.csv")
```

```python
df['label']=df['label'].map({'Fake':0,'Real':1})
print(df.head())
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
labels=df.label
```

```python
labels.head()
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df["Text"],labels,test_size=0.2,random_state=20)
```

```python
x_train.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import PassiveAggressiveClassifier
```

```python
#initilise a Tfidfvectorizer
vector = TfidfVectorizer(stop_words='english',max_df=0.7)
```

```python
#fit and transform
tf_train = vector.fit_transform(x_train)
tf_test=vector.transform(x_test)
```

```python
#initilise a PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=50)
pac.fit(tf_train,y_train)
```

```python
#prediction the test dataset
from sklearn.metrics import accuracy_score, confusion_matrix
y_pred = pac.predict(tf_test)
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy : {round(score*100,2)}%")
```

```python
#confusion metrics
confusion_matrix(y_test, y_pred, labels=['0', '1'])
```

```python
#save model
import pickle
filename = 'finalized_model.pkl'
pickle.dump(pac, open(filename, 'wb'))
```

```python
import joblib
from ipywidgets import Textarea,Button,Output,VBox
from IPython.display import display

#Load Model and vectorizer
model=joblib.load("finalized_model.pkl")
vectorizer=joblib.load("vectorizer.pkl")

#Create UI elements
text_area=Textarea(placeholder="Enter news article text here...",layout={'width':'600px','height':'100px'})
button=Button(description="Check News",button_style='info')
output=Output()

#Define what happens on button click
def on_button_clicked(b):
    news=text_area.value
    if news.strip()=="":
        with output:
            output.clear_output()
            print("Please enter some text")
        return
    X=vectorizer.transform([news])
    prediction=model.predict(X)[0]
    result="Real News" if prediction==1 else "Fake News"
    with output:
        output.clear_output()
        print(f"prediction:{result}")
button.on_click(on_button_clicked)
ui=VBox([text_area,button,output])
display(ui)
```
